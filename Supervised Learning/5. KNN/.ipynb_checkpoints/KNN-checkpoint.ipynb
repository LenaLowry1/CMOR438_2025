# The KNN Model

## Part 1: Data Processing

As usual, begin by importing packages. In this model, we import `train_test_split`, allowing us to randomly partition our data into training and testing data, with our choice of proportion. We will return to using the penguins dataset for this module.
In the chart below, we have plotted the bill length and depth of each of the  3 penguin species. Already, it is easy to see how the data for each species clusters together.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

sns.set_theme()

# Import data
penguins = sns.load_dataset("penguins")

# Plot data
pengs = ["Adelie", "Chinstrap", "Gentoo"]
colors = ["red", "blue", "orange"]

fig, ax = plt.subplots(figsize = (10, 8))
for species, color in zip(pengs, colors):
    temp_df = penguins[penguins.species == species]
    ax.scatter(temp_df.bill_length_mm,
               temp_df.bill_depth_mm,
               c = color,
               label = species, 
               )    
ax.set_xlabel("bill length [mm]", fontsize = 15)
ax.set_ylabel("bill depth [mm]", fontsize = 15)
ax.legend()
plt.show()
```

Next, we will divide our data into training and testing data. Because this dataset is on the larger side, I chose to use 80% for training  and 20% for testing.

```python
X = penguins[["bill_length_mm",
          "bill_depth_mm"]].to_numpy()

y = penguins["species"].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.20, 
                                                    random_state=14)
```

## Part 2: Defining Helper Functions

Further, we will define the Euclidean distance function and test it on a couple of random data points to see its functionality. This function computes the elementwise difference between the two vectors, then computes the dot product of the difference with itself, which equals the sum of squared differences, and finally takes the square root of that sum.

We also define an absolute value distance function as an alternative. This function sums up the absolute value of piecewise distance of the two vectors.

```python
def euc_dist(p, q):
    return np.sqrt((p - q) @ (p - q))

def abs_dist(p, q):
    return np.sum(np.abs(p - q))

print(f"The euclidean distance between point 9 and 72 is {euc_dist(X[9], X[72])}")
print(f"The absolute distance between point 9 and 72 is {abs_dist(X[9], X[72])}")
```

The following fucntion, `k_nearest_nbrs`, will:
- Compute distance between input point and each feature point
- Store point, its label, and its distance to the input
- Add to neighbors list
- Sort all neighbors by their distance (smallest first)
- Return the top-k closest neighbors

```python
def k_nearest_nbrs(point, features, labels, k, dist):
    # Create a list for nbrs and distances
    nbrs = []
    
    for p, label in zip(features,labels):
        d = dist(point, p)
        temp_data = [p, label, d]
        nbrs.append(temp_data)
        
    nbrs.sort(key = lambda x : x[-1])
    
    return nbrs[:k]

# try function with a random data point and k value, and show true label 
print(k_nearest_nbrs(X_test[3], X_train, y_train, 7, abs_dist), "\n")
print(k_nearest_nbrs(X_test[3], X_train, y_train, 7, euc_dist), "\n", y_test[3])
```

## Part 3: Making the Model

Now we define our model, `KNN_Predict`, which calls k_nearest_neighbors() to find the k closest neighbors to the point.
If we are performing classification, then the function extracts the labels of the k neighbors and returns the most frequent label using `max(..., key=labels.count)`.

```python
def KNN_Predict(point, features, labels, k, dist = euc_dist, regression = False):
    
    nbrs = k_nearest_nbrs(point, features, labels, k, dist)
    
    if regression == False:
        labels = [x[1] for x in nbrs]
        return max(labels, key = labels.count)
    
    # if we are performing regression, average all k values for the predicted value    
    else:
        return sum(x[1] for x in nbrs)/k
```

A test of the function:

```python
print(KNN_Predict(X_test[3], X_train, y_train, 7))
print(KNN_Predict(X_test[3], X_train, y_train, 7, abs_dist), "\n",y_test[3])
```

The `classification_error` function computes the classification error rate for k-nearest neighbors and returns the proportion of incorrect predictions (float).

```python
def classification_error(test_X, test_y, train_X, train_y, k, dist = euc_dist):
    
    incorrect = sum(
        KNN_Predict(xi, train_X, train_y, k, dist) != yi
        for xi, yi in zip(test_X, test_y)
    )
    return incorrect / len(test_X)

# runs a simple test with our data
print(classification_error(X_test, y_test, X_train, y_train, 9))
print(classification_error(X_test, y_test, X_train, y_train, 9, abs_dist))
```

Next, we will explore different values of k along with both kinds of distance functions, to see if this will reduce the classification error.

```python
k_values = [k for k in range(1, 40, 2)]

error1 = [classification_error(X_test, y_test, X_train, y_train, k) for k in k_values]
error2 = [classification_error(X_test, y_test, X_train, y_train, k, abs_dist) for k in k_values]

plt.figure(figsize=(10, 8))

# Plot Euclidean distance error
plt.plot(k_values, error1, color='red', marker='o', label='Euclidean Distance')

# Plot Manhattan (abs) distance error
plt.plot(k_values, error2, color='blue', marker='s', label='Absolute Distance')

plt.xlabel('k', fontsize=14)
plt.ylabel('Classification Error', fontsize=14)
plt.title('k-NN Error with Different Distance Metrics', fontsize=16)
plt.xticks(k_values)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

Interestingly, both models acheive their minimum at k = 3, suggesting that it is best to look at only a few close points, rather than a large selection. Looking back at the initial scatter plot of the data, this makes sense. Considering a large number of data points would likely include several data points of different types for points that are on the edge of a cluster, causing the model to missclassify. Generally, both distance functions performed simliarly, although absolute value began with a much higher error.

# Part 4: Regression Problem

Now we will attempt a regression problem. For the feature vector we will use bill length and depth, and we will predict body mass. We will use both distance functions and compare their mse.

```python
# Load and clean the dataset
df = pd.read_csv("palmer_penguins-Copy1.csv")
df_clean = df.dropna(subset=['bill_length_mm', 'bill_depth_mm', 'body_mass_g'])

# Define features and target
X = df_clean[['bill_length_mm', 'bill_depth_mm']].values
y = df_clean['body_mass_g'].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = (y - y.mean()) / y.std()

# Split the data into training and testing sets
train_X, test_X, train_y, test_y = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
```

```python
# Extract variables
bill_length = df_clean['bill_length_mm']
bill_depth = df_clean['bill_depth_mm']
body_mass = df_clean['body_mass_g']

# Create 2D scatter plot with color representing body mass
plt.figure(figsize=(10, 8))
scatter = plt.scatter(bill_length, bill_depth, 
                      c=body_mass, cmap='viridis', edgecolor='k', alpha=0.8)

# Label axes
plt.xlabel("Bill Length (mm)", fontsize=12)
plt.ylabel("Bill Depth (mm)", fontsize=12)
plt.title("Body Mass (color) as a Function of Bill Length and Depth", fontsize=14)

# Color bar to show body mass values
plt.colorbar(scatter, label='Body Mass (g)')

plt.grid(True)
plt.show()
```

For this example, I defined k to be 11, due to the highly clustered nature the data.

```python
k=11

# Generate KNN predictions
y_pred_knn = np.array([KNN_Predict(xi, train_X, train_y, k, regression=True) for xi in test_X])

# Fit a linear regression model between predicted and true values
lr_model = LinearRegression()
y_pred_knn = y_pred_knn.reshape(-1, 1)
test_y = test_y.reshape(-1, 1)
lr_model.fit(y_pred_knn, test_y)

# Predict using the linear model (for plotting the line)
y_line = lr_model.predict(y_pred_knn)

# Plot true vs. predicted with the regression line
plt.figure(figsize=(10, 6))
plt.scatter(y_pred_knn, test_y, color = "lightseagreen", alpha=0.7, label='KNN Predictions vs True')
plt.plot(y_pred_knn, y_line, color='red', label='Linear Fit')
plt.xlabel("KNN Predictions")
plt.ylabel("True Values")
plt.title("Linear Regression KNN Model with Euclidean Distance")
plt.legend()
plt.grid(True)
plt.show()

# Evaluate with MSE
mse = mean_squared_error(test_y, y_pred_knn)
print(f"Mean Squared Error of KNN Predictions: {mse:.4f}")
```

Below we repeat the process but with abs_dist.

```python
y_pred_knn = np.array([KNN_Predict(xi, train_X, train_y, k, abs_dist, regression=True) for xi in test_X])

# Fit a linear regression model between predicted and true values
lr_model = LinearRegression()
y_pred_knn = y_pred_knn.reshape(-1, 1)
test_y = test_y.reshape(-1, 1)
lr_model.fit(y_pred_knn, test_y)

# Predict using the linear model (for plotting the line)
y_line = lr_model.predict(y_pred_knn)

# Plot true vs. predicted with the regression line
plt.figure(figsize=(10, 6))
plt.scatter(y_pred_knn, test_y, color = "lightseagreen", alpha=0.7, label='KNN Predictions vs True')
plt.plot(y_pred_knn, y_line, color='red', label='Linear Fit')
plt.xlabel("KNN Predictions")
plt.ylabel("True Values")
plt.title("Linear Regression KNN Model with Absolute Distance")
plt.legend()
plt.grid(True)
plt.show()

# Evaluate with MSE
mse = mean_squared_error(test_y, y_pred_knn)
print(f"Mean Squared Error of KNN Predictions: {mse:.4f}")
```

From these two tests, we can see that the absolute value function performed slightly better than the euclidean function. Next, we will plot MSE as a function of k.

```python
def knn_mse(test_X, test_y, train_X, train_y, k, dist_fn):
    predictions = [KNN_Predict(xi, train_X, train_y, k, dist_fn) for xi in test_X]
    return mean_squared_error(test_y, predictions)

# Choose a range of k values
k_values = range(1, 75, 2)
mse_scores1 = [knn_mse(test_X, test_y, train_X, train_y, k, euc_dist) for k in k_values]
mse_scores2 = [knn_mse(test_X, test_y, train_X, train_y, k, abs_dist) for k in k_values]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(k_values, mse_scores1, marker='o', color='royalblue', label = "Euclidean")
plt.plot(k_values, mse_scores2, marker='o', color='orange', label = "Absolute")
plt.title('MSE vs. k in KNN Regression', fontsize=14)
plt.xlabel('k (Number of Neighbors)', fontsize=12)
plt.ylabel('Mean Squared Error', fontsize=12)
plt.grid(True)
plt.legend()
plt.xticks(k_values)
plt.show()
```

## Conclusion

- The model performed well for the classification problem, classifying only around 2.9% of examples wrong, regardless of the distance function. Additionally, the optimal value for k was 3, showing that the model does best with information about points only in the target points close vicinity.
- The model did not perform as well for the regresssion problem, at least not for lower values of k (though absolute value seems to perform slightly better as a distance function). We can see in the plot above that the model reached its best performance around k = 51, suggesting that the linear model does better with more context.
