# The Single Neuron Linear Regression Model

## Part 1: Set up
First we import our packages and read in our dataframe.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import seaborn as sns
from sklearn.preprocessing import StandardScaler

sns.set_theme()

df = pd.read_csv("palmer_penguins_copy.csv")
df
````

## Part 2: Data Processing

First, we will inspect our data to see which two features are most strongly correlated. Doing this will ensure we are training our model on data that has a statisitcally significant relationship. I chose to do this by creating a heat map of the correlation coefficients between each numeric variable and every other numeric variable. This visualization allows us to quickly identify a strong choice of variables.

````python
# Removes rows with NaN values
df_clean = df.dropna(subset=["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"])

# Removes the index column, which appears as 'numerical' but is not recorded data
df_clean = df_clean.drop(columns=["Unnamed: 0"])


# Compute correlation matrix
corr = df_clean.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt=".2f", square=True, linewidths=0.5)
plt.title('Correlation Heatmap of Palmer Penguins (Numeric Features)', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Display the plot
plt.tight_layout()
plt.show()
````

Ignoring the diagonals, we can see that flipper length and body mass have the highest correlation coefficient (0.87), making them a promising pair to train our model on. So, we will restict our data frame to those 2 attributes, as well as the first 151 indices (in order to only consider Adelie penguins).

````python
df = pd.read_csv("palmer_penguins_copy.csv")
df = df.iloc[:152][["flipper_length_mm", "body_mass_g"]]
````

Now we plot our data. From the scatter plot, we can see that there **is** a noticiable positive correlation between these variables.

````python
plt.figure(figsize = (10, 8))
plt.scatter(df.flipper_length_mm, 
            df.body_mass_g, 
            color = "lightseagreen",
            label = "Adelie penguins")
plt.xlabel("flipper length [mm]", fontsize = 15)
plt.ylabel("body mass [g]", fontsize = 15)
plt.legend(fontsize = 15)
plt.title("Adelie Penguins Scatter Plot", fontsize = 18)
plt.show()
````

####  Data Formatting
Now we will format our data in preparation for training the model. With this dataset, I found it necessary to standardize the data. Because much of this data consists of larger numbers (in the hundreds or thousdands) which are then being multiplied by random weights or squared by the MSE function, I was recieiving overflow errors with only 5 epochs. To solve this problem, I first removed all NaN terms from the data and then used the ```fit_transform``` method from the```StandardScaler``` class to standardize the X data, and then manually performed the computation for the y data.

````python
X = df.flipper_length_mm.values
X = X.reshape(-1, 1)
y = df.body_mass_g.values

# removes all NaN terms from data
X = X[~np.isnan(X).flatten()]
y = y[~np.isnan(y)]

# normalizes data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = (y - y.mean()) / y.std()
````

Below, we can see the new scatterplot with our modified axes.

````python
plt.figure(figsize = (10, 8))
plt.scatter(X_scaled, y_scaled, color = "lightseagreen", label = "Adelie penguins")
plt.xlabel("flipper length [mm]", fontsize = 15)
plt.ylabel("body mass [g]", fontsize = 15)
plt.title("Normalized Adelie Regression Data", fontsize = 18)
plt.legend(fontsize = 15)
plt.show()
````

## Part 3: Creating and Training the Model
Now we will define our Neuron class. This class implements stochastic graident descent in the ```fit``` method.

````python
class Neuron(object):
    """
    A class representing a single artificial neuron for binary or regression tasks.

    ...

    Attributes
    ----------
    activation_fn : callable
        The activation function applied to the weighted input (e.g., sigmoid, identity, tanh).

    weights_ : numpy.ndarray
        Weight vector including the bias term (last entry). Initialized during training.

    loss_history_ : list
        Stores the mean squared error after each training epoch.

    Methods
    -------
    fit(self, X, y, learning_rate=0.005, epochs=50)
        Trains the neuron using stochastic gradient descent (SGD) on input data X and labels y.

    predict(self, X)
        Computes the neuron's output for the given input(s) X using the learned weights and activation.
    """

    def __init__(self, activation_fn):
        self.activation_fn = activation_fn

    def fit(self, X, y, learning_rate=0.005, epochs=50):
        self.weights_ = np.random.rand(X.shape[1] + 1)  # Last element is bias
        self.loss_history_ = []
        n_samples = X.shape[0]

        for _ in range(epochs):
            total_error = 0
            for x_i, y_i in zip(X, y):
                y_hat = self.predict(x_i)
                error = y_hat - y_i
                self.weights_[:-1] -= learning_rate * error * x_i  # update weights
                self.weights_[-1]  -= learning_rate * error         # update bias
                total_error += 0.5 * error**2
            self.loss_history_.append(total_error / n_samples)
        return self

    def predict(self, X):
        linear_output = np.dot(X, self.weights_[:-1]) + self.weights_[-1]
        return self.activation_fn(linear_output)
````

Next we will define our activiation function (identity function) and train the model.

````python
def linear_activation(z):
    return z

node = Neuron(linear_activation)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
node.fit(X_scaled, y_scaled, learning_rate = 0.0001, epochs = 5)
````

````python
domain = np.linspace(np.min(X_scaled) - .5, np.max(X_scaled) + .5, 100)

plt.figure(figsize = (10, 8))
plt.scatter(X_scaled, y_scaled, color = "lightseagreen", label = "Adelie penguins")
plt.plot(domain, node.predict(domain.reshape(-1, 1)))
plt.xlabel("flipper length [mm]", fontsize = 15)
plt.ylabel("body mass [g]", fontsize = 15)
plt.title("Adelie Regression Data", fontsize = 18)
plt.legend(fontsize = 15)
plt.show()
````

From our plot, we can see that the model appears fairly adept at predicting the body mass. For a more quantitative look at what is happening, we will examine the MSE error over the 5 epochs.

## Part 4: Analysis

````python
plt.figure(figsize = (10, 8))
plt.plot(range(1, len(node.loss_history_) + 1), node.loss_history_, marker = "o",label = "MSE")
plt.xlabel("epochs", fontsize = 15)
plt.ylabel("MSE", fontsize = 15)
plt.xticks(range(1, len(node.loss_history_) + 1))
plt.legend(fontsize = 15)
plt.title("MSE Error at Each Epoch During Training", fontsize = 18)
plt.show()
````

We can see a clear decline in errors over the epochs, showing that the model was effective at learning. Next, we will use many more epochs to see if that provides a more fitting result and reduces errors.

````python
node = Neuron(linear_activation)
node.fit(X_scaled, y_scaled, learning_rate = .0001, epochs = 60)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,10))

ax1.scatter(X_scaled, y_scaled, color = "lightseagreen")
ax1.plot(domain, node.predict(domain.reshape(-1, 1)))
ax1.set_xlabel("flipper length [mm]", fontsize = 15)
ax1.set_ylabel("body mass [g]", fontsize = 15)
ax1.set_title("Adelie Regression Data", fontsize = 18)

ax2.plot(range(1, len(node.loss_history_) + 1), 
         node.loss_history_,
         marker = "o",
         label = "MSE")
ax2.set_xlabel("epochs")
ax2.set_ylabel("MSE")
ax2.set_xticks(range(1, len(node.loss_history_) + 1, 5))
ax2.legend(fontsize = 10)
ax2.set_title("MSE Error at Each Epoch During Training", fontsize = 18)
````

We can see that increasing the epochs allows for more reduction in MSE, although that effect plateaus a bit after 50 epochs.

Finally, we will examine a variety of learning rates. As shown below, the model actually works decently for high learning rate values, such as 0.5. However, around 0.7, nd espcially at 0.75 we see a steep dropoff in the model's predicitive abilities, suggesting that these learning rates are too large and lead to oscillation around the minimum of the MSE.

````python
# Possible learning rates 
rates = [0.01, 0.5, 0.7, 0.75]

domain = np.linspace(np.min(X_scaled) - .5, np.max(X_scaled) + .5, 100)
fig, axs = plt.subplots(2, 2, figsize=(10,10))

# iterate through the axs and rate values
for ax, rate in zip(axs.flat, rates):
    node = Neuron(linear_activation)
    node.fit(X_scaled, y_scaled, learning_rate = rate, epochs = 1_000)
    ax.plot(domain, node.predict(domain.reshape(-1, 1)))
    ax.scatter(X_scaled, y_scaled, color = "lightseagreen")
    ax.set_title(f"learning rate = {rate}", fontsize = 18)
````

## Conclusion

The model succeeded in predicting body mass as a function of flipper length, even with a small number of epochs. We aslo saw reduction in MSE with a greater number of epochs. Experimenting with the learning rate shows that, with this dataset, the model is able to function even with an unusually high learning rate. 

Some advantages of this model are its simplicity and closed form solution, allowing for it to be computationally efficient. However, one contraint is that it assumes all data can be fit linearly, limiting the use cases for the mdoel.
